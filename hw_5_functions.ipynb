```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def al(documents, directories):
    for x in documents:
        for k,v in directories.items():
             for val in v:
                    if x['number'] == val:
                        print('№:', x['number'], 'тип:', x['type'], 'владелец:', x['name'], 'полка:', k )

def owner(documents):
    num = input('Введите номер документа: ')
    if num in [ii for i in directories.values() for ii in i]:
        for x in documents:
            if x['number'] == num:
                dic = x
                own = dic['name']
        print('Владелец документа:', own)
    else:
        print('Документ не найден в базе. Текущий список документов:')
        al(documents, directories)
    return

def schelf(directories):
    num = input('Введите номер документа: ')
    if num in [ii for i in directories.values() for ii in i]:
        for x, y in directories.items():
            if num in y:
                print('Документ хранится на полке:', x)
    else:
        print('Документ не найден в базе. Текущий список документов:')
        al(documents, directories)
    return

def new_schelf(directories):
    new_sch = input('Введите номер новой полки: ')
    schelfs = directories.keys()
    if new_sch not in schelfs:
        directories[new_sch] = None
        print('Полка добавлена. Текущий перечень полок:', *schelfs)
    else:
        print('Такая полка уже существует. Текущий перечень полок:', *schelfs)
    return

def del_schelf(directories):
    del_sch = input('Введите номер полки, которую хотите удалить: ')
    schelfs = directories.keys()
    if del_sch in schelfs:
        if directories[del_sch] == []:
            del directories[del_sch]
            print('Полка удалена. Текущий перечень полок:', *schelfs)
        else:
            print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок:', *schelfs)
    else:
            print('Такой полки не существует. Текущий перечень полок:', *schelfs)
    return

def new_doc(documents, directories):
    n_num = input('Введите номер документа: ')
    n_type = input('Введите тип документа: ')
    n_name = input('Введите владельца документа: ')
    n_sch = input('Введите полку для хранения: ')
    if n_sch in directories.keys():
        documents.append({'type': n_type, 'number': n_num, 'name': n_name})
        for k, v in directories.items():
                if k == n_sch:
                    v.append(n_num)
        print('Документ добавлен. Текущий список документов:')
    else:
        print('Такой полки не существует. Добавьте полку командой as. Текущий список документов:')  
    al(documents, directories)
    return

def del_doc(documents):
    num = input('Введите номер документа: ')
    if num in [ii for i in directories.values() for ii in i]:
        for x in documents:
            if x['number'] == num:
                documents.remove(x)
                for y in directories.values():
                    for yy in y:
                        if yy == num:
                            y.remove(yy)
        print('Документ удален. Текущий список документов:')
    else:
        print('Документ не найден в базе. Текущий список документов:')
    al(documents, directories)  
    return

def change_sch(directories):
    num = input('Введите номер документа: ')
    if num not in [ii for i in directories.values() for ii in i]:
        print('Документ не найден в базе. Текущий список документов:')
        al(documents, directories)  
    else:
        to_schelf = input('Введите номер полки для перемещения: ')
        schelfs = directories.keys()
        if to_schelf not in schelfs:
            print('Такой полки не существует. Текущий перечень полок:', *schelfs)
        else:
            for k, v in directories.items():
                if k == to_schelf:
                    v.append(num)
                elif num in v:
                    v.remove(num)
            print('Документ перемещен. Текущий список документов:')
            al(documents, directories)  
    return

while True:
    commands = ['p', 's','l','ads','ds','ad', 'd', 'm', 'q']
    command = input('Введите команду: ')
    if command not in commands:
        print('Такой команды не существует') 
        command = input('Введите команду: ') 
    if command == 'p':
        owner(documents)
    elif command == 's':
        schelf(directories)
    elif command == 'l':
        al(documents, directories)
    elif command == 'ads':
        new_schelf(directories)
    elif command == 'ds':
        del_schelf(directories)
    elif command == 'ad':
        new_doc(documents, directories)
    elif command == 'd':        
        del_doc(documents)
    elif command == 'm':        
        change_sch(directories)
    elif command =='q':
        break
```
